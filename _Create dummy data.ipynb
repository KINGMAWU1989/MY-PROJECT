```python
import numpy as np
mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
random_number = np.random.multivariate_normal(mean, cov, 500)
print(random_number)
random_number.shape
```

```python
import matplotlib.pyplot as plt
mean = (-3, 0)
mean
cov = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
x,y = np.random.multivariate_normal(mean, cov, 500).T

plt.title('scater')
plt.xlabel("x1")
plt.ylabel("x2")  
plt.scatter(x, y)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("Histogram of x1")


x,y = np.random.multivariate_normal(mean, cov, 500).T
plt.hist(x, color="grey")
plt.xlim(right=3)
plt.show()


plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("Histogram of x2")
plt.show()

x,y = np.random.multivariate_normal(mean, cov, 500).T
plt.hist(x)
plt.xlim(left=-8)
plt.show()
```
